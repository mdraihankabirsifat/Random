```python
import matplotlib.pyplot as plt

# Data from user
dup1_n = [2000,4000,8000,16000,32000,64000,128000,256000]
dup1_t = [1.928,7.378,26.021,94.220,380.297,1540.002,6158.814,24791.132]

dup2_n = [2000,4000,8000,16000,32000,64000,128000,256000]
dup2_t = [0.001551,0.002997,0.005983,0.011998,0.024057,0.048433,0.096326,0.192564]

plt.figure(figsize=(7,5))
plt.loglog(dup1_n, dup1_t, marker='o', label='dup1 (Quadratic)')
plt.loglog(dup2_n, dup2_t, marker='s', label='dup2 (Linear)')

plt.xlabel("Input size N")
plt.ylabel("Time (ms)")
plt.title("Runtime Comparison (log-log)")
plt.legend()
plt.grid(True, which="both", linewidth=0.3)

plt.tight_layout()
plt.show()
```
